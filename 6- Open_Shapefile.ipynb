```python
# Import the packages
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Read the shapefile
ds=gpd.read_file("World_Countries__Generalized_.shp")
```

```python
# Here we can see the attribute table of the shapefile
ds
```

```python
# Plot the shapefile
ds.plot()
```

```python
# Plot the shapefile based on the area of every country and use a color map
ds.plot(column="SHAPE_Area",cmap="OrRd")
```

```python
# Plot the shapefile based on the area of every country and use a color map with legend

fig, ax=plt.subplots(1,1)
ds.plot(column="SHAPE_Area",ax=ax,legend=True)
```

```python

from mpl_toolkits.axes_grid1 import make_axes_locatable
fig, ax=plt.subplots(1,1)
divider=make_axes_locatable(ax)
cax=divider.append_axes("right",size="5%",pad=0.1)
ds.plot(column='SHAPE_Area',ax=ax,legend=True,cax=cax)
```
